## Was sind Ausreißer?

In der Statistik spricht man von einem <b>Ausreißer</b>, wenn ein Messwert oder Befund nicht in eine erwartete Messreihe passt oder allgemein nicht den Erwartungen entspricht. Die „Erwartung“ wird meistens als Streuungsbereich um den Erwartungswert herum definiert, in dem die meisten aller Messwerte zu liegen kommen, z. B. der <b>Quartilabstand</b> Q75 – Q25. Werte, die weiter als das 1,5-Fache des Quartilabstandes außerhalb dieses Intervalls liegen, werden (meist willkürlich) als Ausreißer bezeichnet. Im <b>Boxplot</b> werden besonders hohe Ausreißer gesondert dargestellt. 

### Behandlung von Ausreißer

```python
q75, q50 , q25= np.quantile(dfp['preis'], [.75,.5,.25])
iqr = q75 - q25
dfp=dfp[(dfp['preis']<(q50+1.25*iqr))|(dfp['preis'] <(q50-1.25*iqr))]
q75,q50,q25, (q50-iqr), (q50+iqr)
```

### Preisvergleich

Der Automarkt ist überfluchtet von Fahrzeugen und die Preise schwanken je nach Marke, Modell, Baujahr und Ausstattung unterschiedlich stark. Ein Käufer interessiert sich für den "erwarteten" Preis eines potenziellen Kaufangebots. Basierend auf die Datensätze von Autoscount24 wird nun dieser ermittelt.

```python
dd=pd.read_csv(PATH_PREISVERGLEICH,sep=';',encoding='utf-8')
len(dd)
```

```python
da=dd[(dd['getriebe']=='S') &
      (dd['modell']=='Golf') &
      (dd['karosserieform']=='Limousine') &
      (dd['leistung']==92) &
      (dd['hubraum']==1395)]

b=da.groupby('baujahr').describe()
b.columns = b.columns.get_level_values(1)
b=b.reset_index()

b=b.set_index('baujahr')
b=b.T[:8].T

d=da.groupby('baujahr').mean()
d['preis/km1']=d.preis/d.kilometerstand

s=d.join(b)
s.columns.name="Volkeswagen Golf"
s
```

### Auswertung von Ausstattungen

Wie viel ein Auto wert ist, hängt von vielen Einflussfaktoren ab. Einer der Faktoren sind die Ausstattungen. Dazu gehören zum Beispiel Sommer- und Winterreifen, Sitzheizung, Bi-Xenon Scheinwerfer etc. Mit der folgenden Auswertung kann über das Baujahr die "erwarteten" Ausstattungen der Marke <i>Volkeswagen</i> und Modell <i>Golf</i> ermittelt werden.

```python
dfp = pd.read_csv(PATH_AUSSTATTUNG,sep=';',encoding='utf-8')
dgolf=dfp[(dfp['baujahr']==2016)
      & (dfp['modell']=='Golf')
      & (dfp['karosserieform']=='Limousine')
      & (dfp['leistung']==92)
      ]
len(dgolf)
```

```python
b=['komfort','extras','sicherheit','unterhaltung/media']
extras, l = pyfunc.g(dgolf,b[1])

c = dict(Counter(l))
d={k: c[k] for k in sorted(c,key=c.get, reverse=False)}
plt.figure(figsize=(4, 5))
plt.barh(*zip(*d.items()))
plt.show()
```
